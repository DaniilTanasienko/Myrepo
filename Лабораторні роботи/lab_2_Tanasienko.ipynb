4. Написати процедуру на Python сортування масиву методом бульбашки

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

arr = [24, 54, 25, 18, 22, 21, 60]
print("Вихідний масив:", arr)
bubble_sort(arr)
print("Відсортований масив:", arr)
```
