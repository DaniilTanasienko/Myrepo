### <center> –ó–∞–≤–¥–∞–Ω–Ω—è –¥–ª—è —Å–∞–º–æ—Å—Ç—ñ–π–Ω–æ—ó —Ä–æ–±–æ—Ç–∏ </center>

#### **1)** –ü–∏—à–µ–º–æ —Ñ—É–Ω–∫—Ü—ñ—é pop_n(), —â–æ –≤–∏–¥–∞–ª—è—î –µ–ª–µ–º–µ–Ω—Ç–∏ —Å—Ç–µ–∫–∞ –∑ –π–æ–≥–æ –ø–æ—á–∞—Ç–∫—É –¥–æ –Ω–æ–º–µ—Ä–∞ n –≤–∫–ª—é—á–Ω–æ;

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        return None

    def is_empty(self):
        return len(self.items) == 0

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        return None

    def size(self):
        return len(self.items)

    def pop_n(self, n):
        """
        –í–∏–¥–∞–ª—è—î –µ–ª–µ–º–µ–Ω—Ç–∏ —Å—Ç–µ–∫–∞ –∑ –π–æ–≥–æ –ø–æ—á–∞—Ç–∫—É –¥–æ –Ω–æ–º–µ—Ä–∞ n –≤–∫–ª—é—á–Ω–æ.
        :param n: –Ü–Ω–¥–µ–∫—Å –æ—Å—Ç–∞–Ω–Ω—å–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞ –¥–ª—è –≤–∏–¥–∞–ª–µ–Ω–Ω—è (0-based index).
        """
        if n < 0 or n >= self.size():
            raise IndexError("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω–∏–π —ñ–Ω–¥–µ–∫—Å n.")
        
        # –ü–µ—Ä–µ–≤–µ—Å—Ç–∏ —Å—Ç–µ–∫ —É —Å–ø–∏—Å–æ–∫ —ñ –≤–∏–¥–∞–ª–∏—Ç–∏ –ø–æ—Ç—Ä—ñ–±–Ω—ñ –µ–ª–µ–º–µ–Ω—Ç–∏
        self.items = self.items[n + 1:]

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
stack = Stack()
stack.push(1)
stack.push(2)
stack.push(3)
stack.push(4)
stack.push(5)

print("–ü–æ—á–∞—Ç–∫–æ–≤–∏–π —Å—Ç–µ–∫:", stack.items)
stack.pop_n(2)
print("–ü—ñ—Å–ª—è –≤–∏–¥–∞–ª–µ–Ω–Ω—è –¥–æ —ñ–Ω–¥–µ–∫—Å—É 2 –≤–∫–ª—é—á–Ω–æ:", stack.items)
```

#### **2)** –û—Ü—ñ–Ω—é—î–º–æ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å (—É —Å–µ—Ä–µ–¥–Ω—å–æ–º—É —ñ –≤ –Ω–∞–π–≥—ñ—Ä—à–æ–º—É —Ä–∞–∑—ñ) –ø—Ä–æ—Ü–µ–¥—É—Ä search, insert —ñ delete —Ä–æ–±–æ—Ç–∏ –∑—ñ —Å—Ç–µ–∫–æ–º.

- **–°–∫–ª–∞–¥–Ω—ñ—Å—Ç—å:**
  - –£ —Å–µ—Ä–µ–¥–Ω—å–æ–º—É: $O(n)$, –¥–µ $n$ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —É —Å—Ç–µ–∫—É (–≤ —Å–µ—Ä–µ–¥–Ω—å–æ–º—É –¥–æ–≤–µ–¥–µ—Ç—å—Å—è –ø—Ä–æ–π—Ç–∏ –ø–æ–ª–æ–≤–∏–Ω—É –µ–ª–µ–º–µ–Ω—Ç—ñ–≤).
  - –£ –Ω–∞–π–≥—ñ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É: $O(n)$, —è–∫—â–æ –¥–æ–≤–µ–¥–µ—Ç—å—Å—è –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏ –≤—Å—ñ –µ–ª–µ–º–µ–Ω—Ç–∏.

## –û–ø–µ—Ä–∞—Ü—ñ—è –≤–∏–¥–∞–ª–µ–Ω–Ω—è (delete)
- **–û–ø–∏—Å:** –í–∏–¥–∞–ª–µ–Ω–Ω—è –≤–µ—Ä—Ö–Ω—å–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞ —Å—Ç–µ–∫–∞ (–æ—Å—Ç–∞–Ω–Ω—å–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞ —Å–ø–∏—Å–∫—É).

  - –£ —Å–µ—Ä–µ–¥–Ω—å–æ–º—É: $O(1)$, –±–æ –≤–∏–¥–∞–ª–µ–Ω–Ω—è –æ—Å—Ç–∞–Ω–Ω—å–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞ –≤ —Å–ø–∏—Å–∫—É –∑–∞–π–º–∞—î –ø–æ—Å—Ç—ñ–π–Ω–∏–π —á–∞—Å.
  - –£ –Ω–∞–π–≥—ñ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É: $O(1)$, –∞–¥–∂–µ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –æ–ø–µ—Ä–∞—Ü—ñ—ó –∑–∞–ª–∏—à–∞—î—Ç—å—Å—è –Ω–µ–∑–º—ñ–Ω–Ω–æ—é.


–ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É –±—ñ–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ—à—É–∫—É:
- –ù–∞–π–≥—ñ—Ä—à–∏–π –≤–∏–ø–∞–¥–æ–∫: $O(\log n)$ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ª–µ–º–µ–Ω—Ç—ñ–≤, —è–∫—ñ –ø–æ—Ç—Ä—ñ–±–Ω–æ –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏, –∑–º–µ–Ω—à—É—î—Ç—å—Å—è –≤–¥–≤—ñ—á—ñ –Ω–∞ –∫–æ–∂–Ω—ñ–π —ñ—Ç–µ—Ä–∞—Ü—ñ—ó.
- –ù–∞–π–∫—Ä–∞—â–∏–π –≤–∏–ø–∞–¥–æ–∫: $ùëÇ(1)$ ‚Äî —à—É–∫–∞–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç –∑–Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –Ω–∞ —Å–µ—Ä–µ–¥–Ω—ñ–π –ø–æ–∑–∏—Ü—ñ—ó –Ω–∞ –ø–µ—Ä—à—ñ–π —ñ—Ç–µ—Ä–∞—Ü—ñ—ó.

#### **3)** –ü–∏—à–µ–º–æ —Ñ—É–Ω–∫—Ü—ñ—é print_n(), —â–æ –¥—Ä—É–∫—É—î –µ–ª–µ–º–µ–Ω—Ç–∏ —á–µ—Ä–≥–∏ –∑ –π–æ–≥–æ –ø–æ—á–∞—Ç–∫—É –¥–æ –Ω–æ–º–µ—Ä–∞ n –≤–∫–ª—é—á–Ω–æ;

```python
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, item):
        """–î–æ–¥–∞—î –µ–ª–µ–º–µ–Ω—Ç –¥–æ –∫—ñ–Ω—Ü—è —á–µ—Ä–≥–∏."""
        self.items.append(item)

    def dequeue(self):
        """–í–∏–¥–∞–ª—è—î –µ–ª–µ–º–µ–Ω—Ç –∑ –ø–æ—á–∞—Ç–∫—É —á–µ—Ä–≥–∏."""
        if not self.is_empty():
            return self.items.pop(0)
        return None

    def is_empty(self):
        """–ü–µ—Ä–µ–≤—ñ—Ä—è—î, —á–∏ –ø–æ—Ä–æ–∂–Ω—è —á–µ—Ä–≥–∞."""
        return len(self.items) == 0

    def size(self):
        """–ü–æ–≤–µ—Ä—Ç–∞—î –∫—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —É —á–µ—Ä–∑—ñ."""
        return len(self.items)

    def print_n(self, n):
        """
        –î—Ä—É–∫—É—î –µ–ª–µ–º–µ–Ω—Ç–∏ —á–µ—Ä–≥–∏ –∑ –ø–æ—á–∞—Ç–∫—É –¥–æ –Ω–æ–º–µ—Ä–∞ n –≤–∫–ª—é—á–Ω–æ.
        :param n: –Ü–Ω–¥–µ–∫—Å –æ—Å—Ç–∞–Ω–Ω—å–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞ –¥–ª—è –¥—Ä—É–∫—É (0-based index).
        """
        if n < 0 or n >= self.size():
            raise IndexError("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω–∏–π —ñ–Ω–¥–µ–∫—Å n.")
        
        # –î—Ä—É–∫—É—î–º–æ –µ–ª–µ–º–µ–Ω—Ç–∏ –∑ –ø–æ—á–∞—Ç–∫—É –¥–æ n –≤–∫–ª—é—á–Ω–æ
        for i in range(n + 1):
            print(self.items[i])

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
queue = Queue()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
queue.enqueue(4)
queue.enqueue(5)

print("–ß–µ—Ä–≥–∞:", queue.items)

print("–ï–ª–µ–º–µ–Ω—Ç–∏ —á–µ—Ä–≥–∏ –¥–æ —ñ–Ω–¥–µ–∫—Å—É 2 –≤–∫–ª—é—á–Ω–æ:")
queue.print_n(2)
```

### –û–ø–µ—Ä–∞—Ü—ñ—è –¥–æ–¥–∞–≤–∞–Ω–Ω—è (insert)
- **–û–ø–∏—Å:** –î–æ–¥–∞–≤–∞–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç–∞ –¥–æ –∫—ñ–Ω—Ü—è —á–µ—Ä–≥–∏ –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –º–µ—Ç–æ–¥—É `append()`.
- **–°–∫–ª–∞–¥–Ω—ñ—Å—Ç—å:**

  - –£ –Ω–∞–π–≥—ñ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É: $O(1)$.

## 2. –ß–µ—Ä–≥–∞ —á–µ—Ä–µ–∑ –¥–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω—é —á–µ—Ä–≥—É (deque)
Python –º–æ–¥—É–ª—å `collections.deque` –¥–æ–∑–≤–æ–ª—è—î —Ä–µ–∞–ª—ñ–∑—É–≤–∞—Ç–∏ —á–µ—Ä–≥—É, –æ–ø—Ç–∏–º—ñ–∑–æ–≤–∞–Ω—É –¥–ª—è –¥–æ–¥–∞–≤–∞–Ω–Ω—è —Ç–∞ –≤–∏–¥–∞–ª–µ–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —ñ–∑ –æ–±–æ—Ö –∫—ñ–Ω—Ü—ñ–≤.

### **1)** –©–æ —Ç–∞–∫–µ —Å—Ç–µ–∫ —ñ —è–∫—ñ –æ–ø–µ—Ä–∞—Ü—ñ—ó –º–æ–∂–Ω–∞ –≤–∏–∫–æ–Ω—É–≤–∞—Ç–∏ –∑—ñ —Å—Ç–µ–∫–æ–º?

**–°—Ç–µ–∫** ‚Äî —Ü–µ —Å—Ç—Ä—É–∫—Ç—É—Ä–∞ –¥–∞–Ω–∏—Ö, —â–æ –ø—Ä–∞—Ü—é—î –∑–∞ –ø—Ä–∏–Ω—Ü–∏–ø–æ–º "–æ—Å—Ç–∞–Ω–Ω—ñ–π –ø—Ä–∏–π—à–æ–≤ ‚Äî –ø–µ—Ä—à–∏–π –≤–∏–π—à–æ–≤" (LIFO, **Last In, First Out**). –ï–ª–µ–º–µ–Ω—Ç–∏ –¥–æ–¥–∞—é—Ç—å—Å—è —Ç–∞ –≤–∏–¥–∞–ª—è—é—Ç—å—Å—è –ª–∏—à–µ –∑ –æ–¥–Ω–æ–≥–æ –∫—ñ–Ω—Ü—è ‚Äî **–≤–µ—Ä—Ö—É —Å—Ç–µ–∫–∞**.

#### –û—Å–Ω–æ–≤–Ω—ñ –æ–ø–µ—Ä–∞—Ü—ñ—ó –∑—ñ —Å—Ç–µ–∫–æ–º:
1. **Push**: –¥–æ–¥–∞–≤–∞–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç–∞ –Ω–∞ –≤–µ—Ä—Ö —Å—Ç–µ–∫–∞.
2. **Pop**: –≤–∏–¥–∞–ª–µ–Ω–Ω—è –≤–µ—Ä—Ö–Ω—å–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞ —Å—Ç–µ–∫–∞.
3. **Peek (–∞–±–æ Top)**: –ø–µ—Ä–µ–≥–ª—è–¥ –≤–µ—Ä—Ö–Ω—å–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞ –±–µ–∑ –π–æ–≥–æ –≤–∏–¥–∞–ª–µ–Ω–Ω—è.
4. **IsEmpty**: –ø–µ—Ä–µ–≤—ñ—Ä–∫–∞, —á–∏ –ø–æ—Ä–æ–∂–Ω—ñ–π —Å—Ç–µ–∫.
5. **Size**: –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —É —Å—Ç–µ–∫—É.


### **2)** –Ø–∫–∞ –æ—Å–Ω–æ–≤–Ω–∞ –≤—ñ–¥–º—ñ–Ω–Ω—ñ—Å—Ç—å –º—ñ–∂ —Å—Ç–µ–∫–æ–º —Ç–∞ —á–µ—Ä–≥–æ—é?

–û—Å–Ω–æ–≤–Ω–∞ –≤—ñ–¥–º—ñ–Ω–Ω—ñ—Å—Ç—å –º—ñ–∂ **—Å—Ç–µ–∫–æ–º** —ñ **—á–µ—Ä–≥–æ—é** –ø–æ–ª—è–≥–∞—î –≤ –ø—Ä–∏–Ω—Ü–∏–ø–∞—Ö —ó—Ö —Ä–æ–±–æ—Ç–∏:

1. **–°—Ç–µ–∫** –ø—Ä–∞—Ü—é—î –∑–∞ –ø—Ä–∏–Ω—Ü–∏–ø–æ–º **"–æ—Å—Ç–∞–Ω–Ω—ñ–π –ø—Ä–∏–π—à–æ–≤ ‚Äî –ø–µ—Ä—à–∏–π –≤–∏–π—à–æ–≤" (LIFO, Last In, First Out)**:
   - –ï–ª–µ–º–µ–Ω—Ç–∏ –¥–æ–¥–∞—é—Ç—å—Å—è —ñ –≤–∏–¥–∞–ª—è—é—Ç—å—Å—è –ª–∏—à–µ –∑ –æ–¥–Ω–æ–≥–æ –∫—ñ–Ω—Ü—è ‚Äî **–≤–µ—Ä—Ö—É —Å—Ç–µ–∫–∞**.

2. **–ß–µ—Ä–≥–∞** –ø—Ä–∞—Ü—é—î –∑–∞ –ø—Ä–∏–Ω—Ü–∏–ø–æ–º **"–ø–µ—Ä—à–∏–π –ø—Ä–∏–π—à–æ–≤ ‚Äî –ø–µ—Ä—à–∏–π –≤–∏–π—à–æ–≤" (FIFO, First In, First Out)**:
   - –ï–ª–µ–º–µ–Ω—Ç–∏ –¥–æ–¥–∞—é—Ç—å—Å—è –≤ **–∫—ñ–Ω–µ—Ü—å —á–µ—Ä–≥–∏** —ñ –≤–∏–¥–∞–ª—è—é—Ç—å—Å—è –∑ **–ø–æ—á–∞—Ç–∫—É —á–µ—Ä–≥–∏**.

#### –í—ñ–∑—É–∞–ª—ñ–∑–∞—Ü—ñ—è:
- **–°—Ç–µ–∫**: [1, 2, 3] ‚Üí –¥–æ–¥–∞—î–º–æ –µ–ª–µ–º–µ–Ω—Ç 4 ‚Üí [1, 2, 3, 4] ‚Üí –≤–∏–¥–∞–ª—è—î–º–æ –≤–µ—Ä—Ö–Ω—ñ–π (4) ‚Üí [1, 2, 3].
- **–ß–µ—Ä–≥–∞**: [1, 2, 3] ‚Üí –¥–æ–¥–∞—î–º–æ –µ–ª–µ–º–µ–Ω—Ç 4 ‚Üí [1, 2, 3, 4] ‚Üí –≤–∏–¥–∞–ª—è—î–º–æ –ø–µ—Ä—à–∏–π (1) ‚Üí [2, 3, 4].


#### **3)** –Ø–∫ –≤–∏ –º–æ–∂–µ—Ç–µ —Ä–µ–∞–ª—ñ–∑—É–≤–∞—Ç–∏ —Å—Ç–µ–∫ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –º–∞—Å–∏–≤—É —ñ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –∑–≤‚Äô—è–∑–∞–Ω–æ–≥–æ —Å–ø–∏—Å–∫—É? –Ø–∫—ñ –ø–µ—Ä–µ–≤–∞–≥–∏ —Ç–∞ –Ω–µ–¥–æ–ª—ñ–∫–∏ –∫–æ–∂–Ω–æ–≥–æ –ø—ñ–¥—Ö–æ–¥—É?

# –°—Ç–µ–∫ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –º–∞—Å–∏–≤—É

```python
class StackArray:
    def __init__(self, size):
        self.stack = [None] * size  # —Å—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Å–∏–≤ –∑–∞–¥–∞–Ω–æ–≥–æ —Ä–æ–∑–º—ñ—Ä—É
        self.top = -1  # –≤–∫–∞–∑—É—î–º–æ, —â–æ —Å—Ç–µ–∫ –ø–æ—Ä–æ–∂–Ω—ñ–π

    def push(self, value):
        if self.top == len(self.stack) - 1:
            raise Exception("Stack Overflow")  # –ø–µ—Ä–µ–ø–æ–≤–Ω–µ–Ω–Ω—è
        self.top += 1
        self.stack[self.top] = value

    def pop(self):
        if self.top == -1:
            raise Exception("Stack Underflow")  # —Å—Ç–µ–∫ –ø–æ—Ä–æ–∂–Ω—ñ–π
        value = self.stack[self.top]
        self.top -= 1
        return value

    def peek(self):
        if self.top == -1:
            raise Exception("Stack is empty")
        return self.stack[self.top]

# –¢–µ—Å—Ç—É—î–º–æ —Å—Ç–µ–∫
stack_array = StackArray(3)
stack_array.push(10)
stack_array.push(20)
stack_array.push(30)

print("Top element:", stack_array.peek())  # –ü–æ–∫–∞–∑—É—î –≤–µ—Ä—Ö–Ω—ñ–π –µ–ª–µ–º–µ–Ω—Ç
print("Pop element:", stack_array.pop())  # –í–∏–¥–∞–ª—è—î –≤–µ—Ä—Ö–Ω—ñ–π –µ–ª–µ–º–µ–Ω—Ç
print("Top element after pop:", stack_array.peek())  # –ü–æ–∫–∞–∑—É—î –≤–µ—Ä—Ö–Ω—ñ–π –µ–ª–µ–º–µ–Ω—Ç –ø—ñ—Å–ª—è pop
```

# C—Ç–µ–∫ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –∑–≤‚Äô—è–∑–∞–Ω–æ–≥–æ —Å–ø–∏—Å–∫—É

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class StackLinkedList:
    def __init__(self):
        self.top = None  # —Å—Ç–µ–∫ –ø–æ—á–∞—Ç–∫–æ–≤–æ –ø–æ—Ä–æ–∂–Ω—ñ–π

    def push(self, value):
        new_node = Node(value)
        new_node.next = self.top
        self.top = new_node

    def pop(self):
        if self.top is None:
            raise Exception("Stack Underflow")  # —Å—Ç–µ–∫ –ø–æ—Ä–æ–∂–Ω—ñ–π
        value = self.top.value
        self.top = self.top.next
        return value

    def peek(self):
        if self.top is None:
            raise Exception("Stack is empty")
        return self.top.value

# –¢–µ—Å—Ç—É—î–º–æ —Å—Ç–µ–∫
stack_linked_list = StackLinkedList()
stack_linked_list.push(10)
stack_linked_list.push(20)
stack_linked_list.push(30)

print("Top element:", stack_linked_list.peek())  # –ü–æ–∫–∞–∑—É—î –≤–µ—Ä—Ö–Ω—ñ–π –µ–ª–µ–º–µ–Ω—Ç
print("Pop element:", stack_linked_list.pop())  # –í–∏–¥–∞–ª—è—î –≤–µ—Ä—Ö–Ω—ñ–π –µ–ª–µ–º–µ–Ω—Ç
print("Top element after pop:", stack_linked_list.peek())  # –ü–æ–∫–∞–∑—É—î –≤–µ—Ä—Ö–Ω—ñ–π –µ–ª–µ–º–µ–Ω—Ç –ø—ñ—Å–ª—è pop
```
