# **1)** Вивчити самостійно і записати (будь-яким способом) алгоритм бульбашкового сортування. Оцінити асимптотику алгоритму сортування методом бульбашки в найгіршому і в найкращому випадку. Порівняти за цими показниками бульбашковий алгоритм з алгоритмом сортування вставлянням. Чому на практиці бульбашковий алгоритм виявляється менш ефективним у порівнянні з сортуванням методом зливанням?

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):  # Елементи після n-i вже на своїх місцях
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Приклад використання
arr = [6, 5, 4, 3, 2, 1,]
sorted_arr = bubble_sort(arr)
print("Відсортований масив:", sorted_arr)
```

## Асимптотика:

1. **Найгірший випадок (масив відсортовано у зворотному порядку):**
   - Кількість порівнянь: $ \frac{n(n-1)}{2} $.
   - Асимптотика: $ O(n^2) $.

2. **Найкращий випадок (масив вже відсортовано):**
   - Якщо використовувати оптимізовану версію, яка припиняє виконання, якщо немає обмінів:
     - Кількість порівнянь: $ O(n) $.
   - Якщо алгоритм не оптимізований: $ O(n^2) $.

3. **Середній випадок:**
   - $ O(n^2) $.

---

## Порівняння з алгоритмом сортування вставлянням:

### Алгоритм сортування вставлянням:
- Принцип: масив розділяється на дві частини — відсортовану і невідсортовану. Кожен елемент з невідсортованої частини вставляється у правильну позицію відсортованої частини.

### Асимптотика сортування вставлянням:
1. **Найгірший випадок:** $ O(n^2) $ (масив відсортовано у зворотному порядку).
2. **Найкращий випадок:** $ O(n) $ (масив вже відсортовано).
3. **Середній випадок:** $ O(n^2) $.

### Порівняння:
- У найкращому випадку сортування вставлянням працює швидше ($ O(n) $), тоді як бульбашковий алгоритм без оптимізації залишається $ O(n^2) $.
- Алгоритм вставляння має меншу кількість операцій порівняння та переміщення елементів.

# **2)** Оцінити асимптотичну складність алгоритму сортування зливанням, скориставшись основною теоремою рекурсії

## Оцінка асимптотичної складності алгоритму сортування зливанням

Алгоритм сортування зливанням працює за принципом **"розділяй і владарюй"**: 

1. Масив ділиться навпіл на дві частини.  
2. Кожна частина сортується рекурсивно.  
3. Після сортування підмасиви зливаються в один відсортований масив.

---

## Основна теорема рекурсії

Основна теорема рекурсії використовується для оцінки часу роботи рекурсивних алгоритмів. Вона визначає асимптотичну оцінку розв'язків рекурентних рівнянь вигляду:

$T(n) = a \cdot T\left(\frac{n}{b}\right) + f(n)$,

де:  
- $a$ — кількість підзадач, на які розбивається оригінальна задача.  
- $b$ — фактор зменшення розміру задачі (на скільки разів зменшується розмір на кожному рівні).  
- $f(n)$ — витрати на злиття (робота поза рекурсією).

---

## Рекурентне рівняння для сортування зливанням

Для сортування зливанням:  
- Масив ділиться на дві частини (2 підзадачі): $a = 2$,  
- Кожна підзадача має розмір у 2 рази менший: $b = 2$,  
- Злиття двох підмасивів потребує $f(n) = O(n)$ часу.

Рекурентне рівняння:

$T(n) = 2 \cdot T\left(\frac{n}{2}\right) + O(n)$.

---

## Аналіз основної теореми рекурсії

1. **Обчислимо логарифмічний член**:  
   $ \log_b(a) = \log_2(2) = 1.$  

2. **Порівняємо $f(n)$ із $n^{\log_b(a)}$:**  
   - $f(n) = O(n)$,  
   - $n^{\log_2(2)} = O(n)$.  

Оскільки $f(n)$ зростає з тією ж швидкістю, що й $n^{\log_b(a)}$, то згідно з теоремою рекурсії:

$T(n) = O(n \log n)$.

---

## Висновок

Асимптотична складність алгоритму сортування зливанням:  

$T(n) = O(n \log n)$.  

Це ефективніше, ніж бульбашкове сортування ($O(n^2)$), особливо для великих масивів.


# **3)** Вивчити і записати (будь-яким способом) самостійно алгоритм швидкого сортування. Оцінити асимптотичну складність алгоритму швидкого сортування, скориставшись основною теоремою рекурсії

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[0]  # Вибір опорного елемента
    less = [x for x in arr[1:] if x <= pivot]  # Елементи менші за опорний
    greater = [x for x in arr[1:] if x > pivot]  # Елементи більші за опорний
    return quick_sort(less) + [pivot] + quick_sort(greater)

# Приклад використання
arr = [10, 7, 8, 9, 1, 5]
sorted_arr = quick_sort(arr)
print("Відсортований масив:", sorted_arr)
```

# Алгоритм швидкого сортування (Quick Sort)

Алгоритм швидкого сортування базується на принципі **розділяй і владарюй**. Ось основні етапи алгоритму:

1. **Вибір опорного елемента**. Вибирається опорний елемент (pivot), який буде використовуватися для поділу масиву.
2. **Розподіл елементів**. Масив розділяється на дві частини:
   - Ліва частина містить елементи, менші за опорний елемент.
   - Права частина містить елементи, більші за опорний елемент.
3. **Рекурсивний виклик**. Після поділу масиву на дві частини алгоритм рекурсивно сортує обидві частини.

#### Кроки алгоритму:
1. Вибирається опорний елемент (наприклад, перший або останній елемент масиву).
2. Масив розділяється на два підмасиви, де елементи лівої частини менші за опорний елемент, а правої — більші.
3. Цей процес повторюється рекурсивно для лівої та правої частин масиву.

---

## Асимптотична складність алгоритму швидкого сортування

Алгоритм швидкого сортування має рекурсивну структуру, тому для оцінки його складності використаємо **основну теорему рекурсії**.

#### Рекурентне рівняння:

Алгоритм швидкого сортування розділяє масив на дві частини і сортує кожну частину рекурсивно. Витрати на поділ масиву є $O(n)$, оскільки для кожного елемента масиву необхідно порівняти його з опорним елементом.

Отже, рекурентне рівняння виглядає так:

$$
T(n) = 2T\left(\frac{n}{2}\right) + O(n)
$$

де:
- $a = 2$ — кількість підзадач (дві частини).
- $b = 2$ — кожна частина має розмір, що вдвічі менший за початковий.
- $f(n) = O(n)$ — витрати на поділ масиву.

---

## Аналіз за допомогою основної теореми рекурсії

1. **Обчислюємо логарифмічний член**:
   $$
   \log_b(a) = \log_2(2) = 1.
   $$

2. **Порівняємо $f(n)$ із $n^{\log_b(a)}$**:
   - $f(n) = O(n)$,
   - $n^{\log_2(2)} = O(n)$.

Оскільки $f(n)$ зростає з тією ж швидкістю, що й $n^{\log_b(a)}$, то згідно з основною теоремою рекурсії:

$$
T(n) = O(n \log n).
$$

---

## Висновок

Асимптотична складність алгоритму швидкого сортування:

$$
T(n) = O(n \log n).
$$

Це означає, що в середньому та в найкращому випадку швидке сортування має дуже хорошу ефективність порівняно з іншими алгоритмами, такими як бульбашкове сортування ($O(n^2)$).

Проте в найгіршому випадку (коли масив вже відсортований або відсортований у зворотному порядку) складність може досягати $O(n^2)$, якщо вибір опорного елемента не оптимальний. Для уникнення цього найкраще використовувати варіанти швидкого сортування з вибором опорного елемента за допомогою методу "медіана трьох" або випадковим чином.

### Чому вона важлива?

## Алгоритми сортування з квадратичною складністю

Алгоритми сортування з квадратичною складністю в найгіршому випадку (O(n²)) включають:

1. **Бульбашкове сортування**  
2. **Сортування вставками**  
3. **Сортування вибором**  

### Чому це є проблемою для великих обсягів даних?

У випадку великих обсягів даних (коли n велике), квадратична складність призводить до значних затримок. Для кожного елемента потрібно порівнювати та/або переміщати інші елементи, що робить час виконання алгоритму пропорційним квадрату кількості елементів ($n^2$). 

Наприклад, для масиву з 1000 елементів кількість операцій буде близько 1 мільйона, а для масиву з 10,000 елементів — вже 100 мільйонів, що робить ці алгоритми неефективними для великих обсягів даних.


#### **3)** В чому полягає перевага сортування злиттям над сортуванням вставками для великих наборів даних?


## Перевага сортування злиттям над сортуванням вставками для великих наборів даних

### 1. Асимптотична складність
- **Сортування злиттям** має складність $O(n \log n)$ у найгіршому випадку, що робить його значно ефективнішим для великих наборів даних.
- **Сортування вставками** має квадратичну складність $O(n^2)$ у найгіршому випадку, що робить його повільним для великих масивів.

### 2. Масштабованість
- Сортування злиттям працює значно швидше з великими наборами даних, оскільки кожен крок алгоритму має логарифмічний ефект на кількість операцій.
- Сортування вставками стає непрактичним для великих даних, оскільки кількість порівнянь і переміщень елементів зростає квадратично.

### 3. Постійна кількість операцій при злитті
- В сортуванні злиттям злиття підмасивів завжди займає $O(n)$ часу, незалежно від початкового порядку елементів.
- Сортування вставками потребує більше операцій при кожному вставленні, що робить його малоефективним при великих обсягах даних.

Таким чином, **сортування злиттям** є значно більш ефективним і масштабованим, ніж **сортування вставками**, коли мова йде про обробку великих наборів даних.

## Алгоритми сортування в стандартних бібліотеках мов програмування

### 1. **Python**
   - **Алгоритм**: Python використовує **Timsort**, який є гібридним алгоритмом, поєднуючи сортування злиттям та сортування вставками.
   - **Часова складність**: $O(n \log n)$ у найгіршому випадку.
   - **Особливості**: Timsort оптимізований для реальних даних і використовує структури даних для зменшення кількості порівнянь і переміщень.

### 2. **Java**
   - **Алгоритм**: Java використовує **Merge Sort** для сортування об'єктів (клас `Arrays.sort()` для об'єктів) та **Dual-Pivot Quicksort** для примітивних типів (наприклад, `int[]`).
   - **Часова складність**:
     - **Merge Sort**: $O(n \log n)$ у найгіршому випадку.
     - **Quicksort**: $O(n \log n)$ у середньому випадку, але $O(n^2)$ у найгіршому.
   - **Особливості**: Java використовує алгоритм швидкого сортування для примітивних типів, оскільки він є більш швидким для малих наборів даних.

### 3. **C++**
   - **Алгоритм**: Стандартна бібліотека C++ (STL) використовує **IntroSort**, який є гібридним алгоритмом, що поєднує **QuickSort**, **HeapSort** та **Insertion Sort**.
   - **Часова складність**:
     - **QuickSort**: $O(n \log n)$ у середньому випадку, але $O(n^2)$ у найгіршому.
     - **HeapSort**: $O(n \log n)$ у найгіршому випадку.
     - **Insertion Sort**: використовується для малих підмасивів.
   - **Особливості**: IntroSort вибирає між QuickSort і HeapSort залежно від розміру масиву та кількості рекурсій.

### Висновок
У стандартних бібліотеках різних мов програмування часто використовуються гібридні алгоритми, які поєднують найкращі характеристики різних методів сортування для досягнення високої ефективності та стабільної роботи з різними наборами даних.


#### **5)** Яка різниця між алгоритмами сортування злиттям і швидким сортуванням? У яких випадках краще використовувати кожен з цих алгоритмів?

## Різниця між алгоритмами сортування злиттям і швидким сортуванням

### 1. **Алгоритм сортування злиттям (Merge Sort)**
   - **Принцип**: Розділяй і владарюй. Масив рекурсивно розділяється на дві частини, які потім сортуються та зливаються назад у відсортований масив.
   - **Часова складність**: $O(n \log n)$ у найгіршому, середньому та найкращому випадку.
   - **Просторова складність**: $O(n)$ — потрібна додаткова пам'ять для злиття.
   - **Стабільність**: Так, зберігає порядок однакових елементів.

### 2. **Алгоритм швидкого сортування (Quick Sort)**
   - **Принцип**: Також використовує метод розділяй і владарюй. Масив розділяється навколо опорного елемента (pivot), і рекурсивно сортуються лівий та правий підмасиви.
   - **Часова складність**: 
     - **Найгірший випадок**: $O(n^2)$ (якщо вибір опорного елемента не оптимальний).
     - **Середній та найкращий випадок**: $O(n \log n)$.
   - **Просторова складність**: $O(\log n)$ — потребує менше пам'яті порівняно з сортуванням злиттям.
   - **Стабільність**: Ні, не зберігає порядок однакових елементів.

---

## Коли використовувати кожен алгоритм?

### **Користуватися сортуванням злиттям (Merge Sort)**
   - Коли потрібна **стабільність сортування** (наприклад, при сортуванні за кількома критеріями).
   - Якщо важливо, щоб алгоритм мав гарантовану **складність $O(n \log n)$** в найгіршому випадку, наприклад, при сортуванні великих наборів даних з непередбачуваними властивостями.
   - Підходить для сортування **великих даних, що зберігаються на диску** (запис у зовнішню пам'ять).

### **Користуватися швидким сортуванням (Quick Sort)**
   - Якщо швидкість виконання є критичною, оскільки **швидке сортування** зазвичай працює швидше за сортування злиттям завдяки меншій потребі в пам'яті.
   - Коли не важлива стабільність сортування.
   - Для **молодих або малих наборів даних** швидке сортування є кращим вибором завдяки меншій кількості операцій.
   - Якщо можна **оптимізувати вибір опорного елемента**, щоб уникнути найгіршого випадку.

#### **6)** Які фактори слід враховувати при виборі алгоритму сортування для конкретної задачі?

## Фактори для вибору алгоритму сортування

При виборі алгоритму сортування для конкретної задачі слід враховувати кілька важливих факторів:

### 1. **Розмір вхідних даних**
   - Для **малих наборів даних** часто достатньо простих алгоритмів, таких як **сортування вставками** або **сортування вибором**.
   - Для **великих наборів даних** оптимальніше використовувати більш ефективні алгоритми, такі як **швидке сортування** або **сортування злиттям**, які мають складність $O(n \log n)$.

### 2. **Часова складність**
   - Якщо важлива **максимальна швидкість** сортування, то варто вибирати алгоритм з **складністю $O(n \log n)$** в найгіршому випадку (наприклад, **Merge Sort** або **Quick Sort**).
   - Якщо в найгіршому випадку дозволяється більша складність, можна використовувати **Quick Sort**, однак для гарантованої швидкості без погіршення в найгіршому випадку варто використовувати **Timsort** або **Merge Sort**.

### 3. **Просторова складність**
   - Якщо обмежена **пам'ять**, варто використовувати алгоритми з **малою просторовою складністю**, такі як **Quick Sort**, який має складність $O(\log n)$.
   - Алгоритми, що вимагають додаткової пам'яті, такі як **Merge Sort**, можуть бути непридатними, якщо пам'ять є обмеженим ресурсом.

### 4. **Стабільність сортування**
   - Якщо важливо **зберегти порядок однакових елементів**, слід вибирати **стабільні алгоритми** сортування, такі як **Merge Sort** або **Timsort**.
   - Якщо стабільність не є критичною, можна використовувати алгоритми на зразок **Quick Sort** або **Heap Sort**.

### 5. **Тип даних**
   - Для **сортування примітивних типів даних** (наприклад, чисел) зазвичай підходять **Quick Sort** або **Heap Sort**.
   - Для **структур даних**, таких як списки об'єктів, краще використовувати **Timsort** (в Python) або **Merge Sort**, оскільки вони стабільні.

### 6. **Властивості вхідних даних**
   - Якщо дані **майже відсортовані** або мають багато однакових елементів, можна використовувати **сортування вставками** або **Timsort**, оскільки вони ефективно працюють в таких випадках.
   - Якщо дані **вже відсортовані** або близькі до сортування, можна скористатися алгоритмами з меншими витратами на порівняння, наприклад, **сортування вставками**.

### 7. **Простота реалізації**
   - Якщо необхідна **швидка та проста реалізація**, можна використовувати **Bubble Sort** або **Insertion Sort**, хоча вони будуть менш ефективні для великих наборів даних.
   - **Сортування злиттям** або **Quick Sort** складніші у реалізації, але забезпечують значно кращу продуктивність.

### 8. **Паралельність**
   - Для великих даних, коли важливе **паралельне оброблення**, алгоритми, такі як **Merge Sort**, можуть бути ефективно паралелізовані для розподілених систем, тоді як **Quick Sort** та інші рекурсивні алгоритми також можуть бути паралелізовані, але з обмеженнями.
